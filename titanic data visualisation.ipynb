```python
import numpy as np
```

```python
import pandas as pd 
```

```python
df=pd.read_csv('tested.csv')
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
print(df.shape)
```

```python
print(df.dtypes)
```

```python
print(df.describe())
```

```python
df.info()
```

```python
df.value_counts()
```

```python
print(df.isnull().sum())
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Fare'].fillna(df['Fare'].median(), inplace=True)
```

```python
df['Cabin'].fillna('Unknown', inplace=True)
```

```python
df.head()
```

```python
df['Deck'] = df['Cabin'].apply(lambda x: x[0] if x != 'Unknown' else 'U')
```

```python
print(df.isnull().sum())
```

```python
df.hist(figsize=(12, 10), bins=30, edgecolor='black')
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
categorical_cols = ['Sex', 'Pclass', 'Embarked', 'Deck']

for col in categorical_cols:
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Class')
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(data=df, x='Pclass', y='Age')
plt.title('Age Distribution by Pclass')
plt.show()
```

```python
sns.violinplot(data=df, x='Survived', y='Fare')
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

df[numerical_cols].hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived')
plt.suptitle('Pairplot of Numerical Features Colored by Survival', y=1.02)
plt.show()
```

```python
sns.boxplot(data=df, x='Sex', y='Fare')
plt.title('Fare Distribution by Gender')
plt.show()
```

```python
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', style='Sex')
plt.title('Scatterplot of Age vs Fare by Survival and Sex')
plt.show()
```

```python
sns.scatterplot(data=df, x='SibSp', y='Parch', hue='Survived')
plt.title('Siblings/Spouses vs Parents/Children by Survival')
plt.show()
```
